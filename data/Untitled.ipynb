route_short_name (routes.txt) -> route_id (routes.txt) -> route_id (trips.txt) -> trip_id (trips.txt) -> basing on some trip id we can receive all the -> stop_id (stop_times.txt) 

```python
def get_all_the_stops(bus_number: int) -> tuple(list[str]:
    
    # A is ukrainian within this string
    route_short_name = f'А{str(bus_number)}'
    routes = pd.read_csv('routes.txt')
    
    route_id = routes[routes['route_short_name'] == route_short_name].iloc[0][0]
    
    trips = pd.read_csv('trips.txt')
    
    specific_route_trips = trips[trips['route_id'] == route_id]
    some_trip_id = specific_route_trips.iloc[0][2]
    print(some_trip_id)
    
    stops = pd.read_csv('stop_times.txt')
    
    some_trip = stops[stops['trip_id'] == some_trip_id]
    
    # from some_trip_stops we can also receive arrival_time of each trip
    some_trip_stops = some_trip['stop_id'].tolist()
    print(some_trip_stops)
    
    stops_df = pd.read_csv('stops.txt')

    stop_names = []
    for idx, elm in enumerate(some_trip_stops):
        stop_names.append(stops_df[stops_df['stop_id'] == some_trip_stops[idx]]['stop_desc'].tolist()[0])
        
    return stop_names, some_trip_stops
    
get_all_the_stops(18)
```
